```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df["class"] = cancer.target

df.tail()
```

# Change it to PyTorch Tensor

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# from_numpy() return double tensor
torch.from_numpy(df.values).dtype
```

```python
data = torch.from_numpy(df.values).float()
print(data.dtype, data.shape)
```

```python
x = data[:, :10]
y = data[:, -1:]
print(x.shape, y.shape)
```

```python
# Train / Valid / Test ratio
ratios = [0.6, 0.2, 0.2]

train_cnt = int(data.size(0) * ratios[0])
valid_cnt = int(data.size(0) * ratios[1])
test_cnt = data.size(0) - train_cnt - valid_cnt
cnts = [train_cnt, valid_cnt, test_cnt]

print("Train %d / Valid %d / Test %d samples." % (train_cnt, valid_cnt, test_cnt))
```

```python
indices = torch.randperm(data.size(0))

x = torch.index_select(x, dim=0, index=indices)
y = torch.index_select(y, dim=0, index=indices)

x = x.split(cnts, dim=0)
y = y.split(cnts, dim=0)

for x_i, y_i in zip(x, y):
    print(x_i.size(), y_i.size())
```

```python
n_epochs = 10000
batch_size = 128
print_interval = 500
early_stop = 100
```

# Get DataLoaders

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.labels[idx]
```

```python
train_loader = DataLoader(
    dataset=CustomDataset(x[0], y[0]),
    batch_size=batch_size,
    shuffle=True,  # Allow shuffling only for training set.
)
valid_loader = DataLoader(
    dataset=CustomDataset(x[1], y[1]),
    batch_size=batch_size,
    shuffle=False,
)
test_loader = DataLoader(
    dataset=CustomDataset(x[2], y[2]),
    batch_size=batch_size,
    shuffle=False,
)

print(
    "Train %d / Valid %d / Test %d samples."
    % (
        len(train_loader.dataset),
        len(valid_loader.dataset),
        len(test_loader.dataset),
    )
)
```

```python
model = nn.Sequential(
    nn.Linear(10, 6),
    nn.LeakyReLU(),
    nn.Linear(6, 5),
    nn.LeakyReLU(),
    nn.Linear(5, 4),
    nn.LeakyReLU(),
    nn.Linear(4, 3),
    nn.LeakyReLU(),
    nn.Linear(3, 1),
    nn.Sigmoid(),
)
```

```python
model
```

```python
optimizer = optim.Adam(model.parameters())
```

# Train

```python
from copy import deepcopy

lowest_loss = np.inf
best_model = None

lowest_epoch = np.inf
```

```python
train_loader.dataset[0]
```

```python
train_history, valid_history = [], []

for i in range(n_epochs):
    y_hat = []

    train_loss = 0
    model.train()
    for x_i, y_i in train_loader:
        y_hat_i = model(x_i)
        loss = F.binary_cross_entropy(y_hat_i, y_i)

        optimizer.zero_grad()
        loss.backward()

        optimizer.step()
        train_loss += float(loss)

    train_loss = train_loss / len(train_loader)

    model.eval()
    with torch.no_grad():
        valid_loss = 0

        for x_i, y_i in valid_loader:
            y_hat_i = model(x_i)
            loss = F.binary_cross_entropy(y_hat_i, y_i)

            valid_loss += float(loss)

            y_hat += [y_hat_i]

    valid_loss = valid_loss / len(valid_loader)

    train_history += [train_loss]
    valid_history += [valid_loss]

    if (i + 1) % print_interval == 0:
        print(
            "Epoch %d: train loss=%.4e  valid_loss=%.4e  lowest_loss=%.4e"
            % (
                i + 1,
                train_loss,
                valid_loss,
                lowest_loss,
            )
        )

    if valid_loss <= lowest_loss:
        lowest_loss = valid_loss
        lowest_epoch = i

        best_model = deepcopy(model.state_dict())
    else:
        if early_stop > 0 and lowest_epoch + early_stop < i + 1:
            print("There is no improvement during last %d epochs." % early_stop)
            break

print("The best validation loss from epoch %d: %.4e" % (lowest_epoch + 1, lowest_loss))
model.load_state_dict(best_model)
```

```python
plot_from = 2

plt.figure(figsize=(20, 10))
plt.grid(True)
plt.title("Train / Valid Loss History")
plt.plot(
    range(plot_from, len(train_history)), train_history[plot_from:], label="train loss"
)
plt.plot(
    range(plot_from, len(valid_history)),
    valid_history[plot_from:],
    label="validation loss",
)
plt.legend()
plt.yscale("log")
plt.show()
```

```python
test_loss = 0
y_hat = []

model.eval()
with torch.no_grad():
    for x_i, y_i in test_loader:
        y_hat_i = model(x_i)
        loss = F.binary_cross_entropy(y_hat_i, y_i)

        test_loss += float(loss)  # Gradient is already detached.

        y_hat += [y_hat_i]

test_loss = test_loss / len(test_loader)
y_hat = torch.cat(y_hat, dim=0)

print("Test loss: %.4e" % test_loss)
```

```python
y_hat.shape
```

```python
correct_cnt = (y[2] == (y_hat > 0.5)).sum()
total_cnt = float(y[2].size(0))

print("Test Accuracy: %.4f" % (correct_cnt / total_cnt))
```
